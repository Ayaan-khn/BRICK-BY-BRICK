```python
import pandas as pd
df=pd.read_excel("Book 6.xlsx")
#importing excels sheet here


# ALL FORMULAS TO CALCULATE COLUMNS -->>>>>>>>>>>
df["Avg"] = (df["maths"] + df["science"] + df["english"]) / 3
df["Total"] = df["maths"] + df["science"] + df["english"]

# ALL COLUMNS BY DEFAULT VALUES -->>>>>>>>>>>>>>>
df["attendance_percentage"] = 0
df["Result"] = "Status_Awaiting"
df["Remarks"]= "Status_Awaiting"


"""RESULT LOGIC"""
df.loc[(df["Avg"] >= 40) & (df["attendance_percentage"] >= 75), "Result"] = "Pass"
df.loc[(df["Avg"] < 40) & (df["attendance_percentage"] < 75), "Result"] = "Fail(Low marks,attendance)"
df.loc[(df["Avg"] >= 40) & (df["attendance_percentage"] < 75), "Result"] = "Fail(Low attendance)"
df.loc[(df["Avg"] < 40) & (df["attendance_percentage"] >= 75), "Result"] = "Fail(Low marks)"
"""REMARKS LOGIC"""
df.loc[df["Result"] == "Pass",  "Remarks"]="Cleared"
df.loc[df["Result"] == "Fail(Low marks,attendance)" , "Remarks"]="Attendance + Academic Issue"
df.loc[df["Result"] == "Fail(Low attendance)" , "Remarks"]="Attendance Issue"
df.loc[df["Result"] == "Fail(Low marks)" , "Remarks"]="Academic Issue"

"""USE IT TO FILL ALL EMPTY VALUES"""
df.fillna("NaN")
```
