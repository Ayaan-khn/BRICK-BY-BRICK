```python
import pandas as pd

customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Ayaan', 'John', 'Sara', 'Mike']
})
orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 4],
    'amount': [500, 1200, 900]
})
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "student_id": [1, 2, 3],
    "name": ["Ayaan", "Rahul", "Sara"]
})
df2 = pd.DataFrame({
    "student_id": [1, 2, 3],
    "marks": [85, 90, 78]
})
```

```python
df1 = df1.set_index("student_id")
df2 = df2.set_index("student_id")

result = df1.join(df2)
print(result)
```
